## HISTOGRAM

Histogram represent the data distribution by forming bins along the range of data and then drawing bars to show the number of observations that fall in each bin.

```python
df.describe()
```

```python
bins = [2,3,4,5,6,7,8,9,10]

sns.distplot(df['time_spend_company'],bins=bins)
plt.xticks(bins)
```

```python
# using some parameters

sns.distplot(df['time_spend_company'],bins=bins, kde=False)
```

```python
sns.distplot(df['time_spend_company'], kde=True)
```

```python
sns.distplot(df['time_spend_company'],bins=bins, hist=False)
```

```python
sns.distplot(df['time_spend_company'],bins=bins, rug=True, hist_kws={'color':'yellow','edgecolor':'red','linewidth':3,'alpha':0.5})
```

```python
bins = [2,3,4,5,6,7,8,9,10]

sns.distplot(df['time_spend_company'],bins=bins,color='brown')
plt.xticks(bins)
```

## SCATTER PLOT

Scatter plot is a plots with bar graph and kernel density estimate(KDE).

Each value in a dataset represented by a dot.

```python
titanic_df = sns.load_dataset('titanic')
titanic_df.head()
```

```python
 sns.scatterplot(x ='age',y ='fare',
                 data = titanic_df)
```

```python
# using some parameters

sns.scatterplot(x ='age',y ='fare',
                 data = titanic_df, hue = 'alive')
```

```python
plt.figure(figsize = (12,6))

sns.scatterplot(x = 'age', y = 'fare',
                data = titanic_df, hue = 'alive', style = 'class', palette = 'inferno',alpha = 0.5)
plt.title('Titanic Data Analysis')
```

### Draw scatterplot and lineplot on the same graph

```python
plt.figure(figsize = (12,6))
sns.scatterplot(x = 'age', y = 'fare',
                data = titanic_df, hue = 'alive',
                style = 'class')

sns.lineplot(x = 'age', y = 'fare',
                data = titanic_df, color = 'green')
plt.title('Titanic Data Analysis')
           
```

## Barplot

Bar plot are the type of data visualization used to represent data in the form of rectangular bars.

It shows relationship between numerical and categorical data.

```python
titanic_df.head()
```

```python
sns.barplot(x ='class',y ='fare', data = titanic_df)
```

```python
# using some parameters

sns.barplot(x ='class',y ='fare',
                 data = titanic_df,
            hue = 'sex')
```

```python
sns.barplot(x ='class',y ='fare',
                 data = titanic_df,
            hue = 'sex', palette = 'icefire', orient='v')
```

```python
sns.barplot(x ='class',y ='fare',
                 data = titanic_df,
            hue = 'sex', palette = 'icefire', orient='v',
            ci=100, errwidth=3, errcolor= 'yellow',saturation = 0.9)
```
